```python
#imports
import requests
import pandas as pd
from datetime import datetime
```

```python
#API 1 Companies config
API_KEY = "VQvlf8GEh34n2tKWSWhl7yo36JJx0MjX"
BASE_URL = "https://financialmodelingprep.com/stable"


symbols = ["AAPL", "MSFT", "AMZN", "GOOGL", "META"]
```

```python
#Company Profile

profiles = []

for symbol in symbols:
    url = f"{BASE_URL}/profile?symbol={symbol}&apikey={API_KEY}"
    data = requests.get(url).json()
    
    if isinstance(data, list) and len(data) > 0:
        profiles.append(data[0])

df_profiles_raw = pd.DataFrame(profiles)
df_profiles_raw
```

```python
#Historical Employee Count

employees_history = []

for symbol in symbols:
    url = f"{BASE_URL}/historical-employee-count?symbol={symbol}&apikey={API_KEY}"
    data = requests.get(url).json()
    
    if isinstance(data, list):
        for row in data:
            employees_history.append(row)

df_employees_raw = pd.DataFrame(employees_history)
df_employees_raw
```

```python
#Market Capitalization

market_caps = []

for symbol in symbols:
    url = f"{BASE_URL}/market-capitalization?symbol={symbol}&apikey={API_KEY}"
    data = requests.get(url).json()
    
    if isinstance(data, list) and len(data) > 0:
        market_caps.append(data[0])

df_marketcap_raw = pd.DataFrame(market_caps)
df_marketcap_raw

```
